```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# (https://scikit-learn.org/stable/modules/clustering.html) * Important Documentation *
```

# DATA 1

```python
data = pd.read_hdf("/Users/dhruvkumar/Desktop/cluster_A_data.hf5")
```

```python
data.shape
```

```python
data.columns
```

```python
data
```

```python
plt.scatter(data["x1"],data["x2"]) # we have make a scatter plot of both columns
```

```python
# KMeans
from sklearn.cluster import KMeans # ML model

model = KMeans(n_clusters=2) #Here, we have to specify the number of clusters
model.fit(data)
# Our ML model is now implimented
```

```python
# Let's Check
print(model.labels_) # following (o & 1)'s  tells us about a particular data point could be in first or second cluster
```

```python
# let's make a wild visualization
def display(data,model):
    labels = model.labels_
    
    plt.scatter(data["x1"],data["x2"],c = labels ,cmap='coolwarm',s=10)
```

```python
display(data=data,model = model)
```
